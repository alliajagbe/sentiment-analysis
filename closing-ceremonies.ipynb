# Renaming the Necessary Columns

```python
closing = closing.rename(columns={'Your contact details: Your name': 'name'})
closing = closing.rename(columns={'Your contact details: Your email': 'email'})
closing = closing.rename(columns={'Which Networks Festival sessions did you participate in?:': 'sessions'})
closing = closing.rename(columns={'Can you share a testimonial - one sentence or a quote of what the festival has meant to you?:': 'testimonial'})
closing = closing.rename(columns={'In what ways has the festival impacted you?: 1': 'impact1'})
closing = closing.rename(columns={'In what ways has the festival impacted you?: 2': 'impact2'})
closing = closing.rename(columns={'In what ways has the festival impacted you?: 3': 'impact3'})
closing = closing.rename(columns={'In what ways has the festival impacted you?: 4': 'impact4'})
closing = closing.rename(columns={'In what ways has the festival impacted you?: 5': 'impact5'})
closing = closing.rename(columns={'In what ways has the festival impacted you?: 6': 'impact6'})
closing = closing.rename(columns={'In what ways has the festival impacted you?: 7': 'impact7'})
closing = closing.rename(columns={'Can you share a concrete story of something that happened as a result of your participation?:': 'story'})
closing.head()
```

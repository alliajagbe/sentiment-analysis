```python
import pandas as pd
```

```python
# read the data
education = pd.read_excel('../data/Education Obstacle - opening.xlsx', header=None)
education.head()
```

# Data Cleaning

```python
# deleting the first two rows
education = education.iloc[2:]
```

```python
# assigning new header
new_header = education.iloc[0]
education = education[1:]
education.columns = new_header

education = education.reset_index(drop=True)
education.head()
```

## Renaming the category columns

```python
# renaming the columns
education = education.rename(columns={'I (choose all that apply):: 1': 'first_category'})
education = education.rename(columns={'I (choose all that apply):: 2': 'second_category'})
education = education.rename(columns={'I (choose all that apply):: 3': 'third_category'})
```

```python
education.head()
```

## Renaming the opinion columns

```python
education = education.rename(columns={'When you hear "thriving education ecosystems", what comes to mind?: 1': 'thriving_ecosystems_1'})
education = education.rename(columns={'When you hear "thriving education ecosystems", what comes to mind?: 2': 'thriving_ecosystems_2'})
education = education.rename(columns={'When you hear "thriving education ecosystems", what comes to mind?: 3': 'thriving_ecosystems_3'})
education = education.rename(columns={'When you hear "thriving education ecosystems", what comes to mind?: 4': 'thriving_ecosystems_4'})
```

```python
education.head()
```

## Choosing only the Required Columns

```python
# choosing only the needed columns
education_clean = education[['first_category', 'second_category', 'third_category', 'thriving_ecosystems_1', 'thriving_ecosystems_2', 'thriving_ecosystems_3', 'thriving_ecosystems_4']]
```

```python
education_clean.head()
```

```python
len(education_clean)
```

# Grouping by First Category

```python
# grouping by the first category
first_category = education_clean.groupby('first_category')
for group, group_data in first_category: 
    print("Group: ", group)
    print("Group data: ", group_data)
    print("\n")
```

# Word Frequency Analysis on Each Group in the First Category

## Importing the required libraries

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
import string
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

### Donor Group

```python
text = ''.join(donors['thriving_ecosystems_1'])
```

```python
tokens = word_tokenize(text)
tokens
```

```python
# removing stopwords
stop_words = set(stopwords.words('english'))
tokens = [w for w in tokens if not w in stop_words]
tokens
```

```python
freq_dist = FreqDist(tokens)
for word, frequency in freq_dist.most_common(10):
    print(u'{};{}'.format(word, frequency))
```

# Word Cloud Visualization for Donor Group

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(freq_dist)
```

```python
plt.figure(figsize=(20,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.savefig('donor_group.png')
plt.show()
```

### Movement Group

```python
text = ''.join(movement['thriving_ecosystems_1'])
tokens = word_tokenize(text)
tokens = [w for w in tokens if not w in stop_words]

# removing punctuation
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in tokens]
stripped = [w for w in stripped if w.isalpha()]
stripped
```

```python
stripped = [word for phrase in stripped for word in phrase.split('Collaboration')]
stripped.append('Collaboration')
stripped
```

```python
freq_dist = FreqDist(stripped)
for word, frequency in freq_dist.most_common(20):
    print(u'{};{}'.format(word, frequency))
```

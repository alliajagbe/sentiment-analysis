```python
import pandas as pd
```

```python
# read the data
education = pd.read_excel('../data/Education Obstacle - opening.xlsx', header=None)
education.head()
```

# Data Cleaning

```python
# deleting the first two rows
education = education.iloc[2:]
```

```python
# assigning new header
new_header = education.iloc[0]
education = education[1:]
education.columns = new_header

education = education.reset_index(drop=True)
education.head()
```

## Renaming the category columns

```python
# renaming the columns
education = education.rename(columns={'I (choose all that apply):: 1': 'first_category'})
education = education.rename(columns={'I (choose all that apply):: 2': 'second_category'})
education = education.rename(columns={'I (choose all that apply):: 3': 'third_category'})
```

## Renaming the opinion columns

```python
education = education.rename(columns={'When you hear "thriving education ecosystems", what comes to mind?: 1': 'thriving_ecosystems_1'})
```
